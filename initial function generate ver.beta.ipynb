```python
import numpy as np
def initial_condition_generater(vx,vy,vz,d1,d2,d3,r1,r2,r3,r,a,n,m_max):

    #d1 最內層的地殼密度/或最內區質點的質量
    #d2 中層地殼密度/或中層質點的質量
    #d3 最外側地殼密度/或最外質點的質量
    #r1,r2,r3 各層地殼距離中心距離
    #r 小行星球形分佈半徑
    #a每質點動量交換半徑 默認為碎片半徑  因此我們在下面會帶入  4/3*a^3*np.pi*r^3*d(i)=mass
    #m_max 每片質量
    #n質點個數
    
    
    if  m_max<=4/3*a**3*np.pi*max(d1,d2,d3) :
         print(str(m_max)+' and '+str(4/3*a**3*np.pi*max(d1,d2,d3)))
         print("input maxium mass/density ilegal")
         return

    if r1>=r2 or r2>=r3 or r3>r :
        print("input ilegal !!")
        return
    
    if a<=1 :
        print("particle radius too small!!")
        return
    
    coor_trans = np.zeros((n,3))
    coor_trans[:,0] =r
    coor_trans[:,1] = np.pi
    coor_trans[:,2] = 2*np.pi
    R=np.zeros((n,8))
    R[:,7]=a
    R[:,3]=vx
    R[:,4]=vy
    R[:,5]=vz
    spherical = np.random.rand(n,3)*coor_trans


    position = np.zeros((n,3))
    position[:,0] = spherical[:,0]*np.sin(spherical[:,1])*np.cos(spherical[:,2])
    position[:,1] = spherical[:,0]*np.sin(spherical[:,1])*np.sin(spherical[:,2])
    position[:,2] = spherical[:,0]*np.cos(spherical[:,1])
    R[:,0:3] = position

            
            

    i=1
    while i<=n :
       
        if sum(sum(R[i-1:i,0:3])**2)<=r1**2:
            R[i-1:i,6]=4/3*np.pi*a**3*d1
            i=i+1
            print('地殼一')
            restart=False

        elif r1**2<=sum(sum(R[i-1:i,0:3])**2)<=r2**2:
            R[i-1:i,6]=4/3*np.pi*a**3*d2
            i=i+1
            print("地殼二")
            restart=False
        elif r2**2<=sum(sum(R[i-1:i,0:3])**2)<=r3**3 :
            R[i-1:i,6]=4/3*np.pi*a**3*d3
            i=i+1
            restart=False
            print("地殼三")
        else:
            continue
    
    R[:,0:3] = position + np.ones((n,3))*27450
    print(np.sort(R[:,6]))
    return R
print(initial_condition_generater(10,10,10,10**1.12,10**1.45,10**1.66,50,100,150,200,10,70,10**5.3))
```
